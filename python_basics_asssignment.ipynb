13 What is the difference between for and while loops
- The main difference is that a for loop is best for a predetermined number of iterations, while a while loop is used when the number of iterations is unknown and depends on a condition that changes during the loop's execution. A for loop is often more concise for definite loops (like iterating over a list), whereas a while loop is more flexible for indefinite loops (like waiting for a file to be ready).

14 Describe a scenario where a while loop is more suitable than a for loop.
- A while loop is more suitable than a for loop in scenarios where the number of iterations is not known in advance and the loop's termination depends on a specific condition being met, rather than iterating over a fixed sequence or a predetermined range.

```python
#1  Write a Python program to print "Hello,World!

- print("Hello,World!")

#2  Write a Python program that displays your name and age
-
age = input("Enter age: ")
name = input("Enter name: ")
print("Hi, my name is " + name + ", and I am " + age + " years old.")

#3 Write code to print all the pre-defined keywords in Python using the keyword library
-
def print_python_keywords():
    """Prints all pre-de

#4 Write a program that checks if a given word is a Python keyword.
- import keyword

def is_python_keyword(word):
  """
#  Checks if a given word is a Python keyword.

    Args:
    word: The string to check.

    Returns:
    True if the word is a Python keyword, False otherwise.
    """
    return keyword.iskeyword(word)

# Example usage
word_to_check = input("Enter a word to check: ")

if is_python_keyword(word_to_check):
  print(f"'{word_to_check}' is a Python keyword.")
else:
  print(f"'{word_to_check}' is not a Python keyword.")

#5  Create a list and tuple in Python, and demonstrate how attempting to change an element works differently
for eac
- # Create a list
my_list = [10, 20, 30, 40]
print(f"Original list: {my_list}")

# Attempt to change an element in the list
my_list[1] = 25
print(f"Modified list: {my_list}")

# Create a tuple
my_tuple = (100, 200, 300, 400)
print(f"\nOriginal tuple: {my_tuple}")

# Attempt to change an element in the tuple
try:
    my_tuple[1] = 250
except TypeError as e:
    print(f"Error attempting to modify tuple: {e}")

#6   Write a function to demonstrate the behavior of mutable and immutable arguments
-  ## Function to demonstrate mutable behavior (using a list)
def modify_mutable(my_list_arg):
    print(f"Inside function (before modification): {my_list_arg}")
    # Modifying the list in-place using .append()
    my_list_arg.append(4)
    print(f"Inside function (after modification): {my_list_arg}")

## Function to demonstrate immutable behavior (using an integer)
def modify_immutable(my_int_arg):
    print(f"Inside function (before modification): {my_int_arg}")
    # Reassigning the variable creates a NEW integer object
    my_int_arg += 1
    print(f"Inside function (after modification): {my_int_arg}"

#7 Write a program that performs basic arithmetic operations on two user-input numbers.
- def perform_arithmetic_operations():
    """
         # Performs basic arithmetic operations on two user-input numbers.
         num1 = float(input("Enter the first number: "))
         num2 = float(input("Enter the second number: "))

         print("\n--- Results ---")
         print(f"Addition: {num1} + {num2} = {num1 + num2}")
         print(f"Subtraction: {num1} - {num2} = {num1 - num2}")
         print(f"Multiplication: {num1} * {num2} = {num1 * num2}")

        if num2 != 0:
            print(f"Division: {num1} / {num2} = {num1 / num2}")
        else:
            print("Division by zero is not allowed.")

        except ValueError:
        print("Invalid input. Please enter valid numbers.")
        except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    perform_arithmetic_operations()

 #8 . Write a program to demonstrate the use of logical operators.
 # Example: Logical Operators (AND, OR, NOT) with generic variables
a, b, c = True, False, True

# AND: Both conditions must be True
if a and c:
    print("Both a and c are True (AND condition).")

# OR: At least one condition must be True
if b or c:
    print("Either b or c is True (OR condition).")

# NOT: Reverses the condition
if not b:
    print("b is False (NOT condition).")

#9  Write a Python program to convert user input from string to integer, float, and boolean types
-   def convert_user_input():
    """
    Prompts the user for input and attempts to convert it to integer, float,
    and boolean types, handling potential errors during conversion.
    """
    user_input_str = input("Please enter a value: ")

    # Convert to Integer
    try:
        int_value = int(user_input_str)
        print(f"Converted to Integer: {int_value} (Type: {type(int_value)})")
    except ValueError:
        print(f"Cannot convert '{user_input_str}' to Integer.")

    # Convert to Float
    try:
        float_value = float(user_input_str)
        print(f"Converted to Float: {float_value} (Type: {type(float_value)})")
    except ValueError:
        print(f"Cannot convert '{user_input_str}' to Float.")

    # Convert to Boolean
    # A non-empty string is True, an empty string is False.
    # Specific string values like 'true', 'false', '1', '0' can be handled explicitly.
    boolean_value = bool(user_input_str)
    print(f"Converted to Boolean (based on emptiness): {boolean_value} (Type: {type(boolean_value)})")

    # Optional: More explicit boolean conversion for 'true'/'false' strings
    if user_input_str.lower() == 'true':
        explicit_bool_value = True
        print(f"Explicitly converted 'true' to Boolean: {explicit_bool_value} (Type: {type(explicit_bool_value)})")
    elif user_input_str.lower() == 'false':
        explicit_bool_value = False
        print(f"Explicitly converted 'false' to Boolean: {explicit_bool_value} (Type: {type(explicit_bool_value)})")

       if __name__ == "__main__":
      convert_user_input()

#10   Write code to demonstrate type casting with list elements.
  -    # Original list with mixed data types
original_list = ["10", 25.5, "30", 40, "50.7"]

# Type casting elements to integers
# Using a list comprehension for conciseness
int_list = [int(float(item)) if isinstance(item, str) and '.' in item else int(item) for item in original_list]
print(f"Original list: {original_list}")
print(f"List after casting to int: {int_list}")
print(f"Type of first element in int_list: {type(int_list[0])}\n")

# Type casting elements to floats
float_list = [float(item) for item in original_list]
print(f"Original list: {original_list}")
print(f"List after casting to float: {float_list}")
print(f"Type of first element in float_list: {type(float_list[0])}\n")

# Type casting elements to strings
string_list = [str(item) for item in original_list]
print(f"Original list: {original_list}")
print(f"List after casting to string: {string_list}")
print(f"Type of first element in string_list: {type(string_list[0])}\n")

# Demonstrating type casting within a loop
another_list = ["apple", 123, True, 45.67]
casted_to_string_loop = []
for item in another_list:
    casted_to_string_loop.append(str(item))

print(f"Another list: {another_list}")
print(f"List after casting to string using a loop: {casted_to_string_loop}")
print(f"Type of first element in casted_to_string_loop: {type(casted_to_string_loop[0])}")

# 11 Write a program that checks if a number is positive, negative, or zero
- def check_number_type(number):
  """
  Checks if a given number is positive, negative, or zero.

  Args:
    number: The number to be checked.

  Returns:
    A string indicating whether the number is "Positive", "Negative", or "Zero".
  """
  if number > 0:
    return "Positive"
  elif number < 0:
    return "Negative"
  else:
    return "Zero"

# Example usage:
num1 = 10
print(f"The number {num1} is: {check_number_type(num1)}")

num2 = -5
print(f"The number {num2} is: {check_number_type(num2)}")

num3 = 0
print(f"The number {num3} is: {check_number_type(num3)}")

num4 = -0.5
print(f"The number {num4} is: {check_number_type(num4)}")

num5 = 7.2
print(f"The number {num5} is: {check_number_type(num5)}")

#12 Write a for loop to print numbers from 1 to 10
- for i in range(1, 11):
 print(i)

#13  Write a Python program to find the sum of all even numbers between 1 and 50.
- def sum_of_even_numbers(start, end):
    """
    Calculates the sum of all even numbers within a given range (inclusive).

    Args:
        start (int): The starting number of the range.
        end (int): The ending number of the range.

    Returns:
        int: The sum of all even numbers in the range.
    """
    total_sum = 0
    for number in range(start, end + 1):  # Iterate from start to end (inclusive)
        if number % 2 == 0:  # Check if the number is even
            total_sum += number  # Add even numbers to the total
    return total_sum

#Define the range
 start_range = 1
end_range = 50

# Calculate the sum
result = sum_of_even_numbers(start_range, end_range)

# Print the result
print(f"The sum of all even numbers between {start_range} and {end_range} is: {result}")

#14 4. Write a program to reverse a string using a while loop
 def reverse_string_while_loop(input_string):
  """
  Reverses a given string using a while loop.

  Args:
    input_string: The string to be reversed.

  Returns:
    The reversed string.
  """
  reversed_str = ""
  index = len(input_string) - 1  # Start from the last character

  while index >= 0:
    reversed_str += input_string[index]
    index -= 1  # Move to the previous character

  return reversed_str

# Example usage
original_string = "hello"
reversed_result = reverse_string_while_loop(original_string)
print(f"Original string: {original_string}")
print(f"Reversed string: {reversed_result}")

original_string_2 = "Python"
reversed_result_2 = reverse_string_while_loop(original_string_2)
print(f"Original string: {original_string_2}")
print(f"Reversed string: {reversed_result_2}")

#15 . Write a Python program to calculate the factorial of a number provided by the user using a while loop.
- def calculate_factorial_while(number):
    """
    Calculates the factorial of a non-negative integer using a while loop.

    Args:
        number: The non-negative integer for which to calculate the factorial.

    Returns:
        The factorial of the number if it's non-negative, otherwise an error message.
    """
    if number < 0:
        return "Factorial is not defined for negative numbers."
    elif number == 0:
        return 1  # Factorial of 0 is 1
    else:
        factorial = 1
        i = 1
        while i <= number:
            factorial *= i
            i += 1
        return factorial

# Get input from the user
try:
    num = int(input("Enter a non-negative integer: "))
    result = calculate_factorial_while(num)
    print(f"The factorial of {num} is: {result}")
except ValueError:
    print("Invalid input. Please enter an integer.")



```
